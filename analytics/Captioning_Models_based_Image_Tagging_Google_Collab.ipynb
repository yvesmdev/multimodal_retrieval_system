<h2>Step 2b - Florence Captioning</h2>

```python
device = 'cpu' #'gpu'
model_id = 'microsoft/Florence-2-base'
#load Florence-2 model
model = AutoModelForCausalLM.from_pretrained(model_id, device_map = device, trust_remote_code=True).eval()
processor = AutoProcessor.from_pretrained(model_id, device_map = device, trust_remote_code=True)
```

```python
def generate_labels(task_prompt, image, text_input=None):
    if text_input is None:
        prompt = task_prompt
    else:
        prompt = task_prompt + text_input

    inputs = processor(text=prompt, images=image, return_tensors="pt").to(device)

    generated_ids = model.generate(
      input_ids=inputs["input_ids"],
      pixel_values=inputs["pixel_values"],
      max_new_tokens=1024,
      early_stopping=False,
      do_sample=False,
      num_beams=3,
    )

    generated_text = processor.batch_decode(generated_ids, skip_special_tokens=False)[0]

    output = processor.post_process_generation(
        generated_text,
        task=task_prompt,
        image_size=(image.width, image.height)
    )

    return output
```

```python
img = Image.open(image_filename)
short_task_prompt = '<CAPTION>'
short_caption = list(generate_labels(short_task_prompt, img).values())[0]

long_task_prompt = '<MORE_DETAILED_CAPTION>'
long_caption = list(generate_labels(long_task_prompt, img).values())[0]
long_caption = '\n'.join(textwrap.wrap(long_caption))

```

<h2>Sample Dataset Tagging</h2>

```python
img_df['tagging_vitgpt2'] = ''
img_df['florence_short'] = ''
img_df['florence_long'] = ''
```

```python
Ns = len(img_df)
for i in range(Ns):
    image_name = img_df.iloc[i].file_name
    image_filename = '/content/drive/MyDrive/WORK/Standard Bank/sample_imagesv2/'+image_name
    tag_res = image_to_text(image_filename)
    tag_desc = tag_res[0]['generated_text']
    img_df.at[i,'tagging_vitgpt2'] = tag_desc

    img = Image.open(image_filename)
    short_caption = list(generate_labels(short_task_prompt, img).values())[0]
    long_caption = list(generate_labels(long_task_prompt, img).values())[0]
    long_caption = '\n'.join(textwrap.wrap(long_caption))
    img_df.at[i,'florence_short'] = short_caption
    img_df.at[i,'florence_long'] = long_caption
    print('Tagging %d = %s -> Completion time %d/%d'%(i+1,tag_desc, i+1,Ns),end='\n')
```

```python
#Save results
from datetime import datetime
# Get the current date and time
now = datetime.now()
timestamp_str = now.strftime('%Y_%m_%d_%H_%M_%S')
title_ = 'gpt2_florence_sample_dataset_tagged_%s.csv'%timestamp_str
img_df.to_csv(title_,index=False)
print(title_,'succesfully saved')
```
