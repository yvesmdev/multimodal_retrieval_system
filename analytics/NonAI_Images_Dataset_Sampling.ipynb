<h1>Select True Images Only</h1>

```python
imgdb_true_tag_df = imgdb_tag_df[imgdb_tag_df['label']==0]
print('Number of True Images: ', len(imgdb_true_tag_df))
```

```python
import random

random.seed(1234)#for reproducibility

Ns = 500
id_sample_list = random.sample(range(0,len(imgdb_true_tag_df)-1),Ns)
sample_tag_df = imgdb_true_tag_df.iloc[id_sample_list,:]
sample_tag_df.head()
```

<h1>Sample Copying</h1>

```python
import os
import shutil

SOURCE_DIR = os.path.abspath("../img_datasets")  
DEST_DIR = os.path.abspath("../sample_images") 


image_filenames = sample_tag_df.file_name.tolist()

#Check if the destination directory exists, delete if it does
if os.path.exists(DEST_DIR):
    shutil.rmtree(DEST_DIR)  # Delete the directory and its contents
    print(f"Deleted existing directory: {DEST_DIR}")

#Recreate the directory
os.makedirs(DEST_DIR)
print(f"Created new directory: {DEST_DIR}")

#Copy images from SOURCE_DIR to DEST_DIR
for filename in image_filenames:
    src_path = os.path.join(SOURCE_DIR, filename)
    dest_path = os.path.join(DEST_DIR, filename.replace('train_data/',''))

    if os.path.exists(src_path):  # Check if file exists in the source directory
        shutil.copy(src_path, dest_path)
        print(f"Copied: {filename}  {DEST_DIR}")
    else:
        print(f"File not found: {filename}, skipping...")

print("Image copying process completed.")
```

```python
sample_tag_df['file_name'] = sample_tag_df['file_name'].apply(lambda x: x.replace('train_data/',''))
sample_tag_df.head()
```

```python
sample_tag_df.index = [i for i in range(len(sample_tag_df))]
sample_tag_df[['file_name']].to_csv('../sample_images/sample_images.csv',index=False)
```
